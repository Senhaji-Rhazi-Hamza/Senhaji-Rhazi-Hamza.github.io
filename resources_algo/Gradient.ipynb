## 1- Gradient explanation 

### 1.1 - General idea
The gradient is an operator, denoted sometimes as $\nabla$ others as $\overrightarrow{grad}$, and it applies on a function $f$, and we say gradient of $f$, the gradient is the generalization of the derivative in ${\Bbb R}$ to  ${\Bbb R}^n$

Suppose I have a lighter on my pocket, I take it,position my hand in space at point $X_0^T =(x_0, y_0 ,z_0) $ then light it , the temperature $T(x, y ,z)$ in space arround the lighter is going to be higher, the closer we are to the source of fire $(X_0)$

If we want to carectrize how the temperature evolve around a point for example around $X_0$, this carecterization is given by $\overrightarrow{grad} T_{X_0}$, this vector gives the direction where to go (arround $X_0$) if we expect a temperature increase

And here since $X_0$ is the source of fire, it is the highest temperature so it can't increase because it's an extremum which lead to  $\overrightarrow{grad} T_{X_0} = \overrightarrow{0}$ like in  ${\Bbb R}$, you remember ? 

### 1.2 - More formal idea
Suppose we have a [differentiable](https://en.wikipedia.org/wiki/Differentiable_function) function, defined as 


$f\colon\begin{cases}
{\Bbb E}^n \longrightarrow {\Bbb R}\\
x \longmapsto r
\end{cases}$
Where E is any space where $f$ is defined, and $X^T = (x_1,....,x_n) \in {\Bbb E}^n$, gradient of $f$, is the vector defined as $\overrightarrow{grad}f =  \begin{pmatrix} 
 \frac{\partial f}{\partial x_1} \\
 .\\
 \frac{\partial f}{\partial x_i}\\
.\\
\frac{\partial f}{\partial x_n}
\end{pmatrix}$

Where $ \frac{\partial f}{\partial x_i}$ is defined as :  $ \frac{\partial f}{\partial x_i} =\lim\limits_{dx_i\to 0} \frac{f((x_1,...,x_i + dx_i,...,x_n)) - f((x_1,...,x_i,...,x_n))}{dx_i}$ where $dx_i \in {\Bbb R}$ represent a tiny small elementary value


You can get more details if needed [here](https://en.wikipedia.org/wiki/Gradient)

## 2 - Gradient utility for solving linear systems

We are still tackling linear system solving, and here we will see how we come up with the idea of using the gradient to solve a linear system $AX = b$

Suppose we have a [convex](https://en.wikipedia.org/wiki/Convex_function) and [quadric](https://en.wikipedia.org/wiki/Quadric) function $\phi$, such as

$$\phi(X) = \frac{1}{2}X^TAX - bX + C \tag{1}$$

We will see that solving $\nabla\phi(X) = 0$ is equivalent to solve $AX = b$

Let proof that by Matrix calculation

### 2.1 - Proof

$$
\begin{align}
\phi(X) &= \frac{1}{2}X^TAX - bX \\
\implies&(\phi(X))^{'}= \nabla\phi(X) = \frac{1}{2}(X^TAX)^{'} - (bX)^{'} \\
\implies& \nabla\phi(X) = (\frac{1}{2} )2 AX - b\\
\implies& \nabla\phi(X) = AX - b \\
\implies& \nabla\phi(X^*) = 0 \iff AX^* - b = 0 
\end{align}
$$

## 5 - Gradient Descent

### 5.1 - Gradient calculations

We can compute the gradient by 2 ways, we take the mathematical definition :
* We calculate the expressions of the derivatives, and then we have an expression 

$$\overrightarrow{grad}f(x,y) = \begin{bmatrix}
    . \\
    . 
  \end{bmatrix} $$

* Or we compute the limit given in the definition of the gradient expression (We do this method, where wa can't compute the derivative of the function, we will see it in the next Chapter )

We will choose the 1st method of calculating the derivatives



By computing the gradient according to the definition for both $Z(x, y)$ and $T(x, y)$
we endup with :

$\overrightarrow{grad}T(x, y) = \begin{pmatrix} 
\frac{\partial T(x, y)}{\partial x}  \\
\frac{\partial T(x, y)}{\partial y}  
\end{pmatrix}=\begin{pmatrix} 
x  \\
2y  
\end{pmatrix}
$

and for L

$\overrightarrow{grad}L(x, y) = \begin{pmatrix} 
\frac{\partial L(x, y)}{\partial x}  \\
\frac{\partial L(x, y)}{\partial y}  
\end{pmatrix}=\begin{pmatrix} 
\frac{-2x}{(1 + x^2)^2}  \\
\frac{-2y}{(1 + y^2)^2}  
\end{pmatrix}
$

```python
def gradT(x, y):
    return np.array([x , 2 * y]).reshape(2,1)
    
def gradL(x, y):
    return np.array([(-2 * x )/((1 +x**2)**2), (-2 * y )/((1 +y**2)**2)]).reshape(2,1)
```

### 5.2 - Gradient descent sequence

#### 5.2.1 - Convergence

In order to use gradient descent, we should be sure that we might converge, in order to converge,
the function should be [convex](https://en.wikipedia.org/wiki/Differentiable_function), [differentiable](https://en.wikipedia.org/wiki/Differentiable_function), it's gradient should be [Lipschitz continuous](https://en.wikipedia.org/wiki/Lipschitz_continuity) check the [proof](https://perso.telecom-paristech.fr/rgower/pdf/M2_statistique_optimisation/grad_conv.pdf)

#### 5.2.1 - Iterativ sequence

As $\overrightarrow{grad}f$ is a vector that shows the direction where f is increasing locally, for $\alpha \in  {\Bbb R}^+$ small enought to consider that $f(x + \alpha \overrightarrow{grad}f)$ is still in the same region as  $f(x)$ we will have the following inequations valid :

$$f(x + \alpha \overrightarrow{grad}f) >= f(x)$$ 

and

$$f(x - \alpha \overrightarrow{grad}f) <= f(x)$$

As we want to increase $L(x,y)$ we will move in space following this sequence $x_{n+1} = x_n + \alpha_L \overrightarrow{grad}L$ until we converge to an extremum $X^*$ which is expressed by $\overrightarrow{grad}L(X^*) = 0$

and 

As we want to decrease $T(x, y)$ we will move in space following the same sequence we will just reverse the sign of $\alpha$, because we are trying to go in the opposite direction of growing $x_{n+1} = x_n - \alpha_T \overrightarrow{grad}L$ 

### 5.3 - Drawing Gradient descent 

In this section, we will retake our 2 functions, $L(X), T(X)$ apply the gradient descent on them until convergence, take the sequence "path" constructed during the convergence, plot the sequence path over the equipotential levels for both functions

#### 5.3.1 - Defining gradient descent function

```python
def plot_func(f, X, Cs):
    for C in Cs:
        plt.plot(X ,[f(el, C) for el in X ],"black")
        plt.plot(X ,[-f(el, C) for el in X ],"black")

def gradient_desc(x0, y0, gradf, alpha):
    X0 = np.array([x0, y0]).reshape(2,1)
    Xs = []
    i  = 0
    while (np.linalg.norm(gradf(X0[0], X0[1])) > 1e-3) and (i < 2000):
        i = i + 1
        Xs.append(X0)
        X1 = X0 + alpha * gradf(X0[0], X0[1])
        X0 = X1
        #print(i)
    if (i >= 2000):
        print("The method did'nt converge with the rate :", alpha)
    Xs = np.array(Xs)
    Xs = Xs.reshape(Xs.shape[0], Xs.shape[1])
    return Xs
    
    
```

#### 5.3.2 - Applying gradient Descent for Temprature function $T(X)$

```python
#rate definition
#plt.rcParams['figure.figsize'] = 15,5
alpha_T = 0.2

# levels of energy
Cs = np.linspace(0,2,8)

#Random coordinate space within the considired range
x0 = -1.6
y0 = 1.5

X = np.linspace(-2,2,100)

Xs = gradient_desc(x0, y0, gradT, -alpha_T)

ylim(-2.5,2.5)
xlim(-2.5,2.5)
#plot the function f1 according to the different levels of energy
plot_func(f1, X,Cs)
plt.plot(Xs[:,0],Xs[:,1],"ro-")
print("the method converged in " + str(len(Xs)) + " steps")

    
#construct the sequence 
plt.show()
```

## 6 - Discussing the rate of decent $\alpha$

### 6.1 - Why $\alpha$ is important

In the previous examples, alpha have been choosen carefully as we can converge without problems.

In the gradient descent, since you want to converge to the extremum and you want to do it quickly, alpha determines how big are your steps toward the solution

The gradient tells you how the function evolves locally, if you choose a small alpha, you will converge slowly (small steps), and if you choose a big alpha (big steps), you risk to not anymore being considered locally, so the direction given by the gradient w'ont anymore be relevant(w'ont describe how really the function evolve by this location), and you might never converge or you will go over the solution many times before finding it.

## 7 - Optimization of the descent with "Steepest descent" and  "Conjugate gradient method"

### 7.1 - Introduction

We are happy, we have an iterative method that have an utility for solving linear systems, more efficiently than gauss method, but we have seen, that $\alpha$ is a trouble maker, so it is legitime to ask, how well
we can choose alpha in order to converge in the minimum steps possible, for that we will 3 methods $\textbf{"Steepest descent"}$,  $\textbf{"Conjugate direction method"}$ and  $\textbf{"Conjugate gradient method"}$


### 7.2 - Steepest descent gradient 

#### 7.2.1 - Explanation

The "Steepest descent gradient" consist of moving from one equipotential (level of energy) to a better equipotential (level of energy) where the gradient of the second is perpendicular to the first one.



When we compute gradient at a point $X_0$, it gives us a direction $\overrightarrow{grad}f(X_0)$, the more we go "locally" in this direction, the more the function increase, until being at a point $X_1$ where the function does'nt increase anymore in this direction, it expressed by the fact that $\overrightarrow{grad}f(X_0)$ is tangent to the equipotential and if we compute the gradient again $\overrightarrow{grad}f(X_1)$, it will be perpendicular to the previous one 

Using this reasoning, we can construct an expression of $\alpha$ that at each time has the exact value, that allow us to move from one equipotential, to another one, until we get to the extremum

#### 7.2.2 - Calculations of $\alpha$

the general form of the quadratic : $f(X) =  \frac{1}{2}X^TAX - bX + C \implies \overrightarrow{grad}f(X) = \nabla f(X) = AX - b$

So the sequence
$$\begin{align}
X_{n+1} = X_n - \alpha_n\; \overrightarrow{grad}f(X) \iff &X_{n+1} = X_n - \alpha_n\; (AX_n - b)\\
\iff &X_{n+1} = X_n - \alpha_n\; r_n  \;\;\;\;\; where (r_n = AX_n - b) \\
\iff & AX_{n+1} - AX_n =  -\alpha_n\; Ar_n \\
\iff & (AX_{n+1} - b)   - (AX_n - b) =  -\alpha_n\; Ar_n\\
\iff & r_{n+1}   - r_n =  -\alpha_n\; Ar_n \\
\iff & r_n^T (r_{n+1}   - r_n) =  -\alpha_n\; r_n^TAr_n \\
\implies & - r_n^T r_n =  -\alpha_n\; r_n^TAr_n \;\;\; where (r_{n+1}\perp r_n)\\
\implies &  \boxed{ \alpha_n =  \frac{r_n^T r_n}{r_n^TAr_n}} 
\end{align}$$

The $\alpha_n$ is obtained by construction under the condition $(r_{n+1}\perp r_n)$ and express the size of the step, as the gradient of the next step will be perpendicular to the previous one



#### 7.2.3 - Implementation 

```python
# Function that compute alpha
def get_alpha(A,Xn, b):
    rn = A @ Xn - b
    alpha_n = ((rn.T)@(rn)) /(rn.T@A@rn)
    return np.asscalar(alpha_n)

def gradient_desc_steep(X, A, b, tol=1e-3):
    X0 = X
    Xs = [X0]
    i = 0
    while (np.linalg.norm((A@X0 - b)) > tol) and (i < 2000):
        i = i + 1
        alpha = get_alpha(A,X0, b)
        X1 = X0 - alpha * (A@X0 - b)
        X0 = X1
        Xs.append(X0)
    if (i == 2000):
        print("The method did'nt converge with the rate :", alpha)
    else:
        print("the method converged in " + str(len(Xs) - 1) + " steps")
    return np.array(Xs).reshape(len(Xs), X0.shape[0])
 
```

### 7.3 Conjugate Direction

#### 7.3.1 Introduction 

The conjugate direction method (CD) gives garantee that you will find the solution in $n$ iterations maximum where $n$ is the dimension of $X$.

This method could be undestood as following, where ever you are in the space at the position $X_0$ and you are trying to reach the position $X^*$ "the solution", what separate you from $X^*$ is an error $e_0 = X^* - X_0$ that has n component, if you can guess what is each one of it's component at each iteration, you could find $X^*$ starting from $X_0$ in n steps maximum.

That is what the conjugate direction method is about

#### 7.3.3 Conjugate directions idea :

The steepest descent method has a weakness, it's that we usually go over the same direction several times, before getting to the minimizer/maximizer $X^*$ 

What if we could go in each directions just one time and no more, we would be sure to converge at n iteration maximum.

The error $e_0 = X^* - X_0$ could be expressed in an ortogonal base formed by $\{d_1,d_2,....,d_n\}$,  as $e_0 = \sum\limits_{i = 1}^{i = N} \alpha_i d_i$ 

If at each ith step we could move in the space by $ - \alpha_i d_i$, that means that at each iteration we would eliminate a component from $e_0$, the first step would 
be $e_1 = e_0 - \alpha_0 d_0 =  X^* - (X_0 + \alpha_0 d_0) = X^* - (X_1)$

That would imply that $d_i^Te_{i+1} = 0 
\iff d_i^T(e_{i} - \alpha_i d_i) = 0
\iff \boxed{\alpha_i = \frac{d_i^Te_{i}}{d_i^Td_i}}$

Even if we have the orthogonal base $\{d_1,d_2,....,d_n\}$ when we compute the rate here as we have done in steepest descent, we end up with an expression that contain $e_i$ the trouble maker, we have no acess to $e_i$ if we did, the problem would have been solved, since $e_i =  X^* - X_i$

So we are unable to concritize this idea of orthogonal search in the normal space because of the term $e_i$ but in the stretched space we will be able to concritize this idea of orthogonal search because the term $e_i$ will be replaced by the term $Ae_i$ that we will have acess to.





#### 7.3.4 - A-Orthogonal search

##### 7.3.4.1 Setting the foundation 

So the clever idea here is, since we can't operate the orthogonal search in the normal space because of the unknown term $e_i$, let operate an A-orthogonal search, it means that we will do an orthogonal in the stretched space. and eliminate the components of the error $e_0$ in the stretched space

$p_i$ and $p_j$ are A-Orthogonal $\iff <p_i|Ap_j> = 0$ and $0 = <p_i^TA|p_j>$ for $i,j \in \{1,....,n\}\mid(i \ne j)$ 


So we consider an A-orthogonal base $B_p$ = $\{p_1,.,p_i,.,p_n\}$ for $i \in \{1,....,n\}$


$\textbf{Recapitulation}$

* the general form of the quadratic : $f(X) =  \frac{1}{2}X^TAX - bX + C \rightarrow gradf(X) = \nabla f(X) = AX - b$
* $X^*$ is the minimizer of f, so $\nabla f(X^*) = 0 \rightarrow AX^* = b$
* We define the residual as $r_i = AX_i - b = \nabla f(X_i)$
* We define $ei = X^* - X_i$
* We consider $B_p$ an A-orthogonal base formed by A orthogonal vectors 
* The sequence we consider in order to converge is $\boxed{X_{i + 1} = X_i + \alpha_i p_i}$

We have 2 problems here :
* Determination of $\alpha_i$the exact coefficient that reduce the error at the maximum within the direction $p_i$
* Determination of the directions $p_i$ that form the base $B_p$

##### 7.3.4.2 Determination of $\alpha_i$

* Since $\alpha$ is unknown $X_{i+1}$ can be considered as a function of $\alpha$, since $X_{i + 1}(\alpha) = X_i + \alpha pi$

* Since $X_{i+1}$ is a function, $f(X_{i+1})$ is the compose of 2 functions and can be rewritten as $f\circ X_{i + 1}(\alpha)$ 

* I know that mathematicians might beat me with a bat for i am going to say, but what ever :
  * We know that the derivative of composed functions  $(f\circ g)^{'}(x) = f^{'}\circ g(x) \cdot g^{'}(x)$
  * We also know that the gradient is the generalization of the derivative
  * so  $\frac{dfoX_{i+1}(\alpha)}{d\alpha} = 0 \iff (\nabla f)oX_{i + 1}(\alpha) \cdot \frac{dX_{i+1}(\alpha)}{d\alpha} = 0 \\
 \iff (AX_{i+1} - b)^T (X_i - \alpha p_i)^{'} =0\\
 \iff (A(X_{i+1} - X^*))^T (0 + p_i) = 0 \;\; since \;\; b = AX^*\\
 \iff - (Ae_{i+1})^Tpi = 0 \;\; since \;\;  e_i = X^*  -X_{i+1}  \\
 \iff (A(e_{i} - \alpha_i p_i))^Tp_i = 0\\
 \iff (A e_{i})^T - \alpha_i p_i^TA^Tp_i = 0\\
 \iff \alpha_i = \frac{(A e_{i})^Tpi}{p_i^TAp_i}\\
 \iff \alpha_i = \frac{(A(X^* - X_i))^Tpi}{p_i^TAp_i}\\
 \iff \alpha_i = \frac{-(AX_{i} - b)^Tpi}{p_i^TAp_i}\\
 \iff \boxed{\alpha_i =  \frac{-r_i^Tpi}{p_i^TAp_i}}$ 
 
We d'ont have anymore the trouble maker $e_i$ in the expression of $\boxed{\alpha_i =  \frac{-r_i^Tpi}{p_i^TAp_i}}$ and now, if we have $p_i,\; \alpha_i$ can be easlly computed since we know how to compute $r_i$
 

##### 7.3.4.3 Constructing $B_p$

To construct $B_p$ the A-orthogonal base, we need just a base of $R^n$ and with the [
Gram–Schmidt process](https://en.wikipedia.org/wiki/Gram%E2%80%93Schmidt_process) we can construct an A orthogonal base.

Gram–Schmidt process consist of taking a base $\{v_1,..,v_n\}$ and construct from it an orthogonal base $\{u_1,..,u_n\}$ by appying the process :

$p_i = v_i - \sum\limits_{j \;= \;1}^{j \;= \;i - 1} \frac{<v_i\mid p_j>}{<p_j \mid p_j>}\cdot p_j \iff p_i = v_i - \sum\limits_{j \;= \;1}^{j \;= \;i - 1} \beta_{i,j}\cdot p_j $  where $\beta_{i,j} = \frac{<v_i\mid p_j>}{<p_j \mid p_j>} $

Actually we are removing at each $u_i$ we construct all the projection of $v_i$ on the previous vectors constructed.

To obtain A-orthogonality we just replace the usual scalar product $<x\mid y> $ by $<x\mid A y>$ and for the base  $\{v_1,..,v_n\}$ we just take te orthonormal base of ${\Bbb R}^n$ $\{e_1,..,e_n\}$


##### 7.3.4.4 implementation of Gram–Schmidt process

```python
# vector ei generator
def get_ei(i, n):
    ei = np.zeros(n)
    ei[i] = 1
    return ei.reshape(ei.shape[0],1)

def gram_process(base, A):
    new_base = []
    for vi in base:
        ui = vi - np.sum([((vi.T@ (A@new_base[i])) / (new_base[i].T@(A@new_base[i]))) * new_base[i] for i in range(0,len(new_base))])
        new_base.append(ui.reshape(vi.shape))
    return np.array(new_base)
```

##### 7.3.4.5 Conjugate direction algorithm

```python

def get_alpha_CD(ri, pi,A):
    alpha =  -(ri.T@pi)/(pi.T@(A@pi))
    return alpha

def gradient_desc_CD(X, A, b, Rn, tol=1e-3):
    Bp = gram_process(Rn, A)
    X0 = X
    Xs = [X0]
    for pi in Bp:
        ri = (A@X0 - b)
        alpha = get_alpha_CD(ri,pi, A)
        X1 = X0 + alpha * pi
        X0 = X1
        Xs.append(X0)
    print("the method converged in " + str(len(Xs) - 1) + " steps")
    return np.array(Xs).reshape(len(Xs), X0.shape[0])
 
```

### 7.4 Conjugate Gradient method

#### 7.4.1 Introduction 
And finally after that long road we are here, the conjugate gradient method is the same as the conjugate direction, the only difference is that we construct our base $Bp$ from the residuals $\{r_1,...,r_i,...r_n\}$ as we can get rid off the Gram–Schmidt process that have a complexity of $O_{n^3}$
#### 7.4.2 How Gram–Schmidt process 
To show how Gram–Schmidt process process disapear by choosing to construct $B_p$ from $\{r_i\}$, let us establish some results, that we will use in our demonstration:

Since $B_p$ form a base of A-orthogonal vectors the error $e_0$ can be decomposed in this base as following

$$e_ 0 = \sum\limits_{i= 0}^{i = n -1} a_i p_i$$

And since at each iteration in the $\textbf{CD}$ we eliminate component by component, it make sense to express the term $e_j$ by :
$$e_ j = \sum\limits_{i= j}^{i = n -1} a_i p_i$$

And because $\{p_i\}$ are A orthogonal we can infer the following : 
$$p_j^TAe_ j = \sum\limits_{i= j}^{i = n -1} a_i p_j^TAp_i \iff \boxed{p_j^Tr_ j = 0 \;\; for \;\; i\ne j}  \label{eq:7}
\tag{7}$$  

As seen preivously the Gram.. construction expression in A-orthogonal case is :

$$ 
\begin{align}
&p_i^T = v_i^T - \sum\limits_{j \;= \;1}^{j \;= \;i - 1} \beta_{A_{i,j}} p_j^T \;\; \mid  \;\; \beta_{A_{i,j}} = \frac{<v_i\mid Ap_j>}{<p_j \mid Ap_j>} \\
\iff &  p_i^Tr_j = v_i^Tr_j - \sum\limits_{j \;= \;1}^{j \;= \;i - 1} \beta_{A_{i,j}} p_j^Tr_j \;\; \mid  \;\; \beta_{A_{i,j}} = 
 \begin{cases} 
 \frac{<v_i\mid Ap_j>}{<p_j \mid Ap_j>} \;\; \text{if}\ j < i \\
 0 & otherwise
 \end{cases}\\
\implies &  0 = v_i^Tr_j - 0 \;for \;i\ne j  \;\; \text{deduced from equation (7)} \\
\implies &  \boxed{ri^Tr_j = 0 \;\;for \;i\ne j} \; \text{because $v_i = r_i$ , the base choosen is $\{r_i\}$}   \label{eq:8}
\tag{8}
\end{align}$$



Decomposition of $r_{i+1}$

$$\begin{align}
&r_{i+1} = Ae_{i+1}\\
\iff & r_{i+1}= A(e_{i} + \alpha_i p_i)\\
\iff & \boxed{r_{i+1}= Ae_{i} + \alpha_i Ap_i}   \label{eq:9}
\tag{9}
\end{align}$$




$$\begin{align}
& r_i^Tr_{j+1} = r_i^T (r_j + \alpha_j A p_i) \\
\iff& r_i^Tr_{j+1} = r_i^T r_j + r_i^T \alpha_j A p_i) \\
\implies & \boxed{r_i^T \alpha_j A p_i =  \begin{cases} 
\ne 0 & if \text{$i = j + 1$ or $i = j$}  \\
 0 & otherwise
 \end{cases}}\label{eq:10}
\tag{10}\\
\end{align}$$

##### 7.4.2.2 Simplifying $\beta_{A_{i,j}}$
$$\begin{align}
\beta_{A_{i,j}} = 
 \begin{cases} 
 \frac{<v_i\mid Ap_j>}{<p_j \mid Ap_j>} \;\; \text{if}\ j < i \\
 0 \;\; otherwise
 \end{cases} \text{and result (10)} \implies
 \beta_{A_{i,j}} = 
 \begin{cases} 
 \frac{1}{\alpha_{i - 1}} \frac{ri^Tri}{p_{i - 1}^TAp_{i - 1}} \;\; \text{if}\ j = i + 1 \\
 0 & otherwise
 \end{cases}\\
\end{align}$$

We moved from an expression $ \beta_{A_{i,j}}$ to an expression $ \beta_{A_{i,i-1}}$ that depend on i, more than that if we replace $\alpha_{i - 1}$ by it's expression $\alpha_{i - 1} = \frac{-r_{i - 1}^Tp_{i - 1}}{p_{i - 1}^TAp_{i - 1}}$ we end up with  a final expression for $\beta_{A_i}$ such as $\boxed{\beta_{A_i}= \frac{r_{i}^Tr_{i}}{r_{i - 1}^Tr_{i - 1}}} \label{eq:11}
\tag{11} $

It have been painful to achieve this result, but i am really happy that i did, and a big thanks to the [paper](https://www.cs.cmu.edu/~quake-papers/painless-conjugate-gradient.pdf) of [Jonathan_Shewchuk](https://fr.wikipedia.org/wiki/Jonathan_Shewchuk)

##### 7.4.3 Implementing Conjugate gradient method


```python
def get_beta(r0, r1):
    return (r1.T@r1)/(r0.T@r0)
    
def gradient_desc_CG(X, A, b, tol=1e-3):
    X0 = X
    Xs = []
    r0 = (A@X0 - b)
    pk = r0
    while(np.linalg.norm(r0) > tol):
        alphak = get_alpha_CD(r0,pk, A)
        X1 = X0 + alphak * pk 
        r1 = r0 + alphak * A@pk
        betak = get_beta(r0,r1)
        pk = r1 + betak * pk
        Xs.append(X0)
        X0 = X1
        r0 = r1
    Xs.append(X0)
    print("the method converged in " + str(len(Xs) - 1) + " steps")
    return np.array(Xs).reshape(len(Xs), X0.shape[0])
 
```
