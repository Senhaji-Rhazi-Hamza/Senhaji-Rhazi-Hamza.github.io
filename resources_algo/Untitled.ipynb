```python
#### 2.1.2 - Heavy way calculation proof

$$\phi(X) = \frac{1}{2}X^TAX - bX $$

$
AX =  \begin{pmatrix} 
 \sum\limits_{j = 1}^{n}\, a_{1j}x_j\\
 .\\
 \sum\limits_{j = 1}^{n}\, a_{ij}x_j\\
.\\
\sum\limits_{j = 1}^{n}\, a_{nj}x_j
\end{pmatrix} \implies \, X^TAX =   \sum\limits_{i = 1}^{n}(xi\sum\limits_{j = 1}^{n}\, a_{ij}x_j)
$


$bX = \sum\limits_{i = 1}^n \, b_ix_i \label{eq:2}\tag{2}$

From equation (1) and (2): 

$
\begin{align}
\phi(X) &= \frac{1}{2}\sum\limits_{i = 1}^{n}(xi\sum\limits_{j = 1}^{n}\, a_{ij}x_j) \, - \,  \sum\limits_{i = 1}^n \, b_ix_i \\
\iff \phi(X)&= \frac{1}{2}\sum\limits_{i = 1}^{n}x_i\left(a_{ik}x_k + \sum\limits_{\substack{i=1 \\ i\neq k}}^{n}\, a_{ij}x_j)\right)\, - \,  \sum\limits_{\substack{i=1 \\ i\neq k}}^n \, b_ix_i - b_kx_k \\
\iff \phi(X)&=  \frac{1}{2}\sum\limits_{i = 1}^{n}x_i\left(a_{ik}x_k + \sum\limits_{\substack{i=1 \\ i\neq k}}^{n}\, a_{ij}x_j)\right)\, - \,  \sum\limits_{\substack{i=1 \\ i\neq k}}^n \, b_ix_i - b_kx_k\\
\end{align}
$

```

```python
def gradT(x, y):
    return np.array([x , 2 * y]).reshape(2,1)
    
def gradZ(x, y):
    return np.array([(-2 * x )/((1 +x**2)**2), (-2 * y )/((1 +y**2)**2)]).reshape(2,1)

def plot_func(f, X, Cs):
    for C in Cs:
        plt.plot(X ,[f(el, C) for el in X ],"black")
        plt.plot(X ,[-f(el, C) for el in X ],"black")

def gradient_desc(x0, y0, gradf, alpha):
    X0 = np.array([x0, y0]).reshape(2,1)
    Xs = []
    i  = 0
    while (np.linalg.norm(gradf(X0[0], X0[1])) > 1e-3) and (i < 2000):
        i = i + 1
        Xs.append(X0)
        X1 = X0 + alpha * gradf(X0[0], X0[1])
        X0 = X1
        #print(i)
    if (i >= 2000):
        print("The method did'nt converge with the rate :", alpha)
    Xs = np.array(Xs)
    Xs = Xs.reshape(Xs.shape[0], Xs.shape[1])
    return Xs
    
```

```python
#rate definition
#plt.rcParams['figure.figsize'] = 15,5
# y = +-sqrt(( (1 + x**2)/((1 + x**2) * C)) - 1)
def f2(x, C = 1):
    num = 1 + x**2
    den = (1 + x**2) * C - 1
    
    if (num/den - 1) >= 0 :
        return np.sqrt(num/den - 1)
    else:
        return np.nan
alpha_Z = 0.8

# levels of energy
Cs = np.linspace(0,2,8)

#Random coordinate space within the considired range
x0 = -2
y0 = 1.6



Xs = gradient_desc(x0, y0, gradZ, alpha_Z)
X = np.linspace(-5,5,100)

ylim(-5,5)
xlim(-5,5)
#plot the function f1 according to the different levels of energy
plot_func(f2, X,Cs)
plt.plot(Xs[:,0],Xs[:,1],"ro-")
print("the method converged in " + str(len(Xs)) + " steps")
    
#construct the sequence 
plt.show()
```
